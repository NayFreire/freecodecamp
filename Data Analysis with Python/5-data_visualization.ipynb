## Data Visualization

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

Example 1 of plotting with matplotlib:

````python
x = np.arange(-10, 11)
x
````

````python
plt.figure(figsize=(12, 6))

plt.title('My First Plot')

plt.plot(x, x**2) # first plot x = x and y = x^2
plt.plot(x, -1*(x**2)) # second plot x = x and y = -1*(x^2)
````

Example 2 of plotting with matplotlib:

We are going to create 2 plots in one figure, so we need to specify which plot we are working with. To create multiple plots we'll use subplot(). This method uses the following parameters:
- Number of rows: number of rows in grid of the subplot. It determines how the subplots are organized vertically. 
- Number of columns: the number of columns in the grid of the subplot. It determines how the subplots are organized horizontally.
- Index of the subplot: index of the subplot that we are working with in the moment. It starts with 1 as the first plot.

Ex.: in a subplot with 2 rows and 3 columns, the plots would be organized like: 
| 1 | 2 | 3 |
|---|---|---|
| 4 | 5 | 6 |

The line in bold is the fist line, which, in code, would be referred as ```plt.subplot(2, 3, 1)```

````python
plt.figure(figsize=(12,6)) # creating the canvas for the plots
plt.title('My Second Plot')

plt.subplot(1, 2, 1) # working with the first plot
plt.plot(x, x**2)
plt.plot([0, 0, 0], [-10, 0, 100]) 

plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2) # working with the second plot
plt.plot(x, -1*(x**2))
plt.plot([-10, 0, 10], [-50, -50, -50])

plt.legend(['-X^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X Squared')
````

- ```plt.plot([0, 0, 0], [-10, 0, 100])``` creates a vertical line in the middle of the parabolic graphic that goes through the following points (0, -10), (0, 0) and (0, 100)

- ```plt.plot([-10, 0, 10], [-50, -50, -50])``` creates a horizontal line in the middle of the parabolic graphic that goes through the following points (-10, -50), (0, -50), (10, -50)

### Using OOP Interface

In the context of Matplotlib's pyplot interface, Object-Oriented Programming (OOP) is an alternative approach to using the interface compared to the more common stateful interface (using functions like plt.plot(), plt.xlabel(), etc.).

With the OOP approach, you create Figure and Axes objects explicitly and then call methods on these objects to create and customize your plots. Here's a breakdown of how it works:

1. Figure and Axes Objects:
- A Figure object represents the whole figure or window where your plots are drawn. It's like a blank canvas.
- An Axes object represents an individual plot within the figure. It's where you plot your data and customize the plot's appearance.
2. Creating Figure and Axes:
- You create a Figure object using plt.figure().
- You create an Axes object (or multiple Axes objects) using fig.add_subplot() or fig.subplots().
3. Plotting Data:
- Once you have your Axes object, you can plot data on it using methods like plot(), scatter(), bar(), etc., directly on the Axes object.
4. Customizing the Plot:
- You can customize various aspects of the plot, such as labels, titles, legends, colors, etc., by calling methods directly on the Axes object.

````python
# 2. Creating Figure and Axes
fig, axes = plt.subplots(figsize=(12, 6))
````

````python
# 3. Plotting Data
axes.plot(
    x, (x**2),
    color='red', linewidth=3, marker='o', 
    markersize=8, label='X^2'
)
axes.plot(x, -1*(x**2), 'b--', label='-x^2')

axes.set_xlabel('X')
axes.set_ylabel('X squared')

axes.set_title('My First OOP Interface Plot')
axes.legend()
fig
````

````python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x, linestyle='solid')
axes.plot(x, x+1, linestyle='dashed')
axes.plot(x, x+2, linestyle='dashdot')
axes.plot(x, x+3, linestyle='dotted')

axes.set_title('My Second OOP Interface Plot')
````

````python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x, '-og', label="solid green")
axes.plot(x, x+1, '--c', label="dashed cyan")
axes.plot(x, x+2, '-.b', label="dashdot blue")
axes.plot(x, x+3, ':r', label="dotted red")

axes.set_title("My Third OOP Interface Plot")

axes.legend()
````

### Other types of plots

#### Figures and subfigures

When we call the ```subplots()``` function we get a tuple containing a Figure and a axes element.

````python
plot_objects = plt.subplots()

fig, ax = plot_objects

ax.plot([1,2,3], [1,2,3])

plot_objects
````

We can also define how many elements we want inside our figure. To do that we can set the `nrows` and `ncols` params.

````python
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects

plot_objects
````

````python
ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)
ax4.plot(np.random.randn(50), c='yellow')

fig
````

##### The ```subplot2grid()```

Creating a grid-like format in a different way

````python
plt.figure(figsize=(14, 6))
# In a 3x3 subplot, put this plot in the row=0 and col=0, occupying 3 columns
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot(np.random.randn(50), c='red', linestyle='--')

# In a 3x3 subplot, put this plot in the row=1 and col=0, occupying 2 columns
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax2.plot(np.random.randn(50), c='yellow')

# In a 3x3 subplot, put this plot in the row=1 and col=2, occupying 2 rows
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax3.plot(np.random.randn(50), c='green', linestyle=':')

# In a 3x3 subplot, put this plot in the row=1 and col=0
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax4.plot(np.random.randn(50), c='black', linestyle='-.')


# In a 3x3 subplot, put this plot in the row=2 and col=1
ax5 = plt.subplot2grid((3, 3), (2, 1))
ax5.plot(np.random.randn(50), c='orange', linestyle='dashdot')
````

#### Scatter Plot

````python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N)) ** 2

plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=.5, cmap='Spectral')
plt.colorbar()

plt.show()
````

````python
fig = plt.figure(figsize=(14, 6))

# Adding a subplot in the first position
ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

# Adding a subplot in the second position
ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
````

#### Histograms

````python
values = np.random.randn(1000)
````

````python
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=100, alpha=.8, histtype='bar',
         color='steelblue', edgecolor='green')
plt.xlim(xmin=-5, xmax=5)

plt.show()
````

#### KDE (Kernel Density Estimation)

````python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
````

````python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
````

#### Combining plots

````python
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=100, alpha=0.8, density=1,
          histtype='bar', color='steelblue',
          edgecolor='green')

plt.plot(values2, density(values2), color='#FF7F00', linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)

plt.show()
````

#### Bar Plots

````python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
````

````python
plt.figure(figsize=(12, 4))

barwidth = .5
plt.bar(np.arange(len(Y)), Y, width=barwidth, color='#313131')
plt.show()
````

Also can be stacked bars, and add a legend to the plot:

````python
plt.figure(figsize=(12, 4))

barWidth = .5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#313131', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#525252', bottom=Y, label='Label Y2')

plt.legend()
plt.show()
````
