### Getting the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('btc-market-price.csv', header=None) # header = none -> no column names
```

#### Setting the columns names

```python
df.columns = ['Timestamp', 'Price']
df
```

#### Setting Timestamp to datetime type

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.head(3)
```

#### Setting Timestamp as index

```python
df.set_index('Timestamp', inplace=True)
df.loc['2017-04-02']
```

#### Doing all the settings inside .read_csv()

```python
df_btc = pd.read_csv(
    'btc-market-price.csv',
    header=None,
    names=['Timestamp', 'Price'],
    index_col=0,
    parse_dates=True
)

df_eth = pd.read_csv('eth-price.csv', parse_dates=True, index_col=0)
```

### Plotting

```python
df_btc.plot() #using pandas integrated with matplotlib
```

```python
plt.plot(df_btc.index, df_btc['Price']) #using matplotlib
```

```python
prices = pd.DataFrame({'Bitcoin': df_btc['Price'], 
                       'Ether': df_eth['Value']}, 
                       index=df_btc.index)

prices.info()
```

```python
prices.plot(figsize=(12,6))
```

The plot of Ether has a gap, let's see it closer

```python
prices.loc['2017-12-01':'2018-01-01'].plot(figsize=(12, 6))
```
